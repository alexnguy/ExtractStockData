```python
import yfinance as yf
import pandas as pd
```

```python
# We will be using Tesla

tesla = yf.Ticker("TSLA")
```

```python
#Stock info

tesla_info = tesla.info
tesla_info
```

```python
tesla_info["country"]
```

```python
#Extracting share price, ooptions are 1d,5d,1mo,3mo,6mo,1y,2y,5y,10,ytd,and max.

tesla_share_price_data = tesla.history(period="max")
tesla_share_price_data
```

```python
tesla_share_price_data.head()
```

```python
#Reset the index of data frame

tesla_share_price_data.reset_index(inplace=True)
```

```python
# Plot the open price against the date

tesla_share_price_data.plot(x="Date", y="Open")
```
